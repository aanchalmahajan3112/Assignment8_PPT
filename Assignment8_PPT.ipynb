```python
# ðŸ’¡ **Question 1**

# Given two strings s1 and s2, return *the lowest **ASCII** sum of deleted characters to make two strings equal*.

# **Example 1:**

# **Input:** s1 = "sea", s2 = "eat"

# **Output:** 231

# **Explanation:** Deleting "s" from "sea" adds the ASCII value of "s" (115) to the sum.

# Deleting "t" from "eat" adds 116 to the sum.

# At the end, both strings are equal, and 115 + 116 = 231 is the minimum sum possible to achieve this.



def minimumDeleteSum(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(m - 1, -1, -1):
        dp[i][n] = dp[i+1][n] + ord(s1[i])

    for j in range(n - 1, -1, -1):
        dp[m][j] = dp[m][j+1] + ord(s2[j])

    for i in range(m - 1, -1, -1):
        for j in range(n - 1, -1, -1):
            if s1[i] == s2[j]:
                dp[i][j] = dp[i+1][j+1]
            else:
                dp[i][j] = min(ord(s1[i]) + dp[i+1][j], ord(s2[j]) + dp[i][j+1])

    return dp[0][0]


# Example usage
s1 = "sea"
s2 = "eat"
result = minimumDeleteSum(s1, s2)
print(result)
```

```python
# ðŸ’¡ **Question 2**

# Given a string s containing only three types of characters: '(', ')' and '*', return true *if* s *is **valid***.

# The following rules define a **valid** string:

# - Any left parenthesis '(' must have a corresponding right parenthesis ')'.
# - Any right parenthesis ')' must have a corresponding left parenthesis '('.
# - Left parenthesis '(' must go before the corresponding right parenthesis ')'.
# - '*' could be treated as a single right parenthesis ')' or a single left parenthesis '(' or an empty string "".

# **Example 1:**

# **Input:** s = "()"

# **Output:**

# true



def isValid(s):
    open_brackets = []
    asterisks = []
    unbalanced = []

    for i, c in enumerate(s):
        if c == '(':
            open_brackets.append(i)
        elif c == ')':
            if open_brackets:
                open_brackets.pop()
            elif asterisks:
                asterisks.pop()
            else:
                unbalanced.append(i)
        elif c == '*':
            asterisks.append(i)

    while open_brackets and unbalanced:
        if open_brackets[-1] < unbalanced[-1]:
            open_brackets.pop()
            unbalanced.pop()
        else:
            break

    while asterisks and unbalanced:
        if asterisks[-1] > unbalanced[-1]:
            asterisks.pop()
            unbalanced.pop()
        else:
            break

    return len(open_brackets) == 0


# Example usage
s = "()"
result = isValid(s)
print(result)
```

```python

# ðŸ’¡ **Question 3**

# Given two strings word1 and word2, return *the minimum number of **steps** required to make* word1 *and* word2 *the same*.

# In one **step**, you can delete exactly one character in either string.

# **Example 1:**

# **Input:** word1 = "sea", word2 = "eat"

# **Output:** 2

# **Explanation:** You need one step to make "sea" to "ea" and another step to make "eat" to "ea".



def minDistance(word1, word2):
    m, n = len(word1), len(word2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Initialize the first row and column
    for i in range(m + 1):
        dp[i][0] = i
    for j in range(n + 1):
        dp[0][j] = j

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if word1[i - 1] == word2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = min(dp[i - 1][j] + 1, dp[i][j - 1] + 1)

    return dp[m][n]

word1 = "sea"
word2 = "eat"
minDistance(word1, word2)
```

```python

# ðŸ’¡ **Question 4**

# You need to construct a binary tree from a string consisting of parenthesis and integers.

# The whole input represents a binary tree. It contains an integer followed by zero, one or two pairs of parenthesis. The integer represents the root's value and a pair of parenthesis contains a child binary tree with the same structure.
# You always start to construct the **left** child node of the parent first if it exists.

# **Input:** s = "4(2(3)(1))(6(5))"

# **Output:** [4,2,6,3,1,5]



class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def constructTree(s):
    if not s:
        return None

    # Find the index of the first opening parenthesis
    i = 0
    while i < len(s) and s[i] != '(':
        i += 1

    # Extract the root value
    root_val = int(s[:i])

    # Create the root node
    root = TreeNode(root_val)

    # Check if there are any children
    if i < len(s):
        count = 0
        j = i
        while j < len(s):
            if s[j] == '(':
                count += 1
            elif s[j] == ')':
                count -= 1

            if count == 0:
                break
            j += 1

        # Construct the left child
        if i + 1 < j:
            root.left = constructTree(s[i + 1 : j])

        # Construct the right child
        if j + 1 < len(s):
            root.right = constructTree(s[j + 2 : -1])

    return root


def inorderTraversal(root):
    result = []

    def traverse(node):
        if node:
            traverse(node.left)
            result.append(node.val)
            traverse(node.right)

    traverse(root)
    return result


# Example usage
s = "4(2(3)(1))(6(5))"
root = constructTree(s)

# Get the inorder traversal of the constructed tree
inorder = inorderTraversal(root)

print(inorder)
```

```python
# ðŸ’¡ **Question 5**

# Given an array of characters chars, compress it using the following algorithm:

# Begin with an empty string s. For each group of **consecutive repeating characters** in chars:

# - If the group's length is 1, append the character to s.
# - Otherwise, append the character followed by the group's length.

# The compressed string s **should not be returned separately**, but instead, be stored **in the input character array chars**. Note that group lengths that are 10 or longer will be split into multiple characters in chars.

# After you are done **modifying the input array,** return *the new length of the array*.

# You must write an algorithm that uses only constant extra space.

# **Example 1:**

# **Input:** chars = ["a","a","b","b","c","c","c"]

# **Output:** Return 6, and the first 6 characters of the input array should be: ["a","2","b","2","c","3"]

# **Explanation:**

# The groups are "aa", "bb", and "ccc". This compresses to "a2b2c3".



def compress(chars):
    # Pointer to track the current position in the modified array
    pos = 0
    # Pointer to track the current character being processed
    i = 0

    while i < len(chars):
        # Current character
        curr_char = chars[i]
        # Count of consecutive repeating characters
        count = 0

        # Count the number of consecutive repeating characters
        while i < len(chars) and chars[i] == curr_char:
            count += 1
            i += 1

        # Update the character at the current position with the current character
        chars[pos] = curr_char
        pos += 1

        # If the count is greater than 1, update the array with the count
        if count > 1:
            count_str = str(count)
            for c in count_str:
                chars[pos] = c
                pos += 1

    return pos

# Example usage
chars = ["a", "a", "b", "b", "c", "c", "c"]
compressed_length = compress(chars)

# Print the modified array with the compressed characters
print(chars[:compressed_length])
```

```python
# ðŸ’¡ **Question 6**

# Given two strings s and p, return *an array of all the start indices of* p*'s anagrams in* s. You may return the answer in **any order**.

# An **Anagram** is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

# **Example 1:**

# **Input:** s = "cbaebabacd", p = "abc"

# **Output:** [0,6]

# **Explanation:**

# The substring with start index = 0 is "cba", which is an anagram of "abc".

# The substring with start index = 6 is "bac", which is an anagram of "abc".




from collections import Counter

def findAnagrams(s, p):
    # Lengths of the input strings
    n = len(s)
    m = len(p)
    
    # Counter for the pattern string p
    p_counter = Counter(p)
    
    # Counter for the sliding window in s
    window_counter = Counter(s[:m])
    
    # List to store the start indices of anagrams
    indices = []
    
    # Check if the first window is an anagram of p
    if window_counter == p_counter:
        indices.append(0)
    
    # Slide the window across s
    for i in range(1, n - m + 1):
        # Remove the leftmost character from the window
        left_char = s[i - 1]
        if window_counter[left_char] > 1:
            window_counter[left_char] -= 1
        else:
            del window_counter[left_char]
        
        # Add the rightmost character to the window
        right_char = s[i + m - 1]
        window_counter[right_char] += 1
        
        # Check if the window is an anagram of p
        if window_counter == p_counter:
            indices.append(i)
    
    return indices

# Example usage
s = "cbaebabacd"
p = "abc"
anagram_indices = findAnagrams(s, p)

# Print the start indices of p's anagrams in s
print(anagram_indices)
```

```python

# ðŸ’¡ **Question 7**

# Given an encoded string, return its decoded string.

# The encoding rule is: k[encoded_string], where the encoded_string inside the square brackets is being repeated exactly k times. Note that k is guaranteed to be a positive integer.

# You may assume that the input string is always valid; there are no extra white spaces, square brackets are well-formed, etc. Furthermore, you may assume that the original data does not contain any digits and that digits are only for those repeat numbers, k. For example, there will not be input like 3a or 2[4].

# The test cases are generated so that the length of the output will never exceed 105.

# **Example 1:**

# **Input:** s = "3[a]2[bc]"

# **Output:** "aaabcbc"


def decodeString(s):
    stack = []
    current_string = ""
    current_number = 0

    for char in s:
        if char.isdigit():
            # Calculate the number
            current_number = current_number * 10 + int(char)
        elif char == "[":
            # Push the current string and number to the stack
            stack.append(current_string)
            stack.append(current_number)
            # Reset the current string and number
            current_string = ""
            current_number = 0
        elif char == "]":
            # Decode the substring
            number = stack.pop()
            prev_string = stack.pop()
            current_string = prev_string + number * current_string
        else:
            # Append the character to the current string
            current_string += char

    return current_string

# Example usage
s = "3[a]2[bc]"
decoded_string = decodeString(s)

# Print the decoded string
print(decoded_string)
```

```python

# ðŸ’¡ **Question 8**

# Given two strings s and goal, return true *if you can swap two letters in* s *so the result is equal to* goal*, otherwise, return* false*.*

# Swapping letters is defined as taking two indices i and j (0-indexed) such that i != j and swapping the characters at s[i] and s[j].

# - For example, swapping at indices 0 and 2 in "abcd" results in "cbad".

# **Example 1:**

# **Input:** s = "ab", goal = "ba"

# **Output:** true

# **Explanation:** You can swap s[0] = 'a' and s[1] = 'b' to get "ba", which is equal to goal.



def buddyStrings(s, goal):
    # Lengths of the input strings
    n = len(s)
    m = len(goal)
    
    # Check if the lengths of s and goal are equal
    if n != m:
        return False
    
    # Check if s and goal are equal
    if s == goal:
        # Check if s contains at least one repeated character
        seen = set()
        for char in s:
            if char in seen:
                return True
            seen.add(char)
        return False
    
    # Find the mismatched indices
    indices = []
    for i in range(n):
        if s[i] != goal[i]:
            indices.append(i)
    
    # Check if exactly two indices are mismatched
    if len(indices) != 2:
        return False
    
    # Check if swapping the characters at the mismatched indices makes s equal to goal
    i, j = indices
    return s[i] == goal[j] and s[j] == goal[i]

# Example usage
s = "ab"
goal = "ba"
can_swap = buddyStrings(s, goal)

# Print if s and goal can be made equal by swapping two letters
print(can_swap)
```
